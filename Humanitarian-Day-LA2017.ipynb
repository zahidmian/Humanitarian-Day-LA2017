```r
library(ggplot2)
```

```r
dataset <- read.csv('HumanitarianDay.csv')
```

```r
#str(dataset)
```

## Is your homelessness temporary?
### For most it is temporary, but it's not clear what temporary means because later we find that many of them have been homeless for more than 6 months.

```r
ggplot(data = dataset, aes(x = Temporary)) + 
    geom_bar() + 
    geom_text(stat = 'count', aes(label = ..count..), vjust = -1) +
    labs(title = "Is your homelessness temporary", subtitle = "Is your homelessness temporary?") +
    xlab("Temporary (n/a, no, yes)") +
    ylab("Count")
```

## By Gender

## Length of homelessness
### Most have been homeless for less than 2 years, but for as long as 17 years.

```r
ggplot(dataset, aes(x = Length, fill=Gender)) + 
    geom_dotplot(stackgroups = TRUE, method="histodot") +
    scale_x_continuous(breaks = seq(0,240,12)) +
    labs(title = "Length of homelessness") +
    xlab("Length (months)") +
    ylab("Count")
```
